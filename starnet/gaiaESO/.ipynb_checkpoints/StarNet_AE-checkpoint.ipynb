```python
denormalization_path = '/home/spiffical/data/spiffical/gaia-ESO/mu_std_INTRIGOSS_gaiaeso_UVES-4835-5395_7labels.txt'
def denormalize(lb_norm):
    with open(denormalization_path,'r') as f1:
        mu = np.array(map(float, f1.readline().split()[0:7]))
        sigma = np.array(map(float, f1.readline().split()[0:7]))
    return ((lb_norm*sigma)+mu)
```

```python
indx_highT = denormalize(y_synth_train)[:,0] > 5500
indx_lowT = denormalize(y_synth_train)[:,0] <= 5500
X_synth_highT = x_synth_train[indx_highT]
X_synth_lowT = x_synth_train[indx_lowT]
y_synth_highT = y_synth_train[indx_highT]
y_synth_lowT = y_synth_train[indx_lowT]
```
