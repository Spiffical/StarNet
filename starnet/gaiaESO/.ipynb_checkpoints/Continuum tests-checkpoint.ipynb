```python
# Borrowed from https://github.com/radio-astro-tools/statcont/blob/master/statcont/cont_finding.py
# Modified by Spencer Bialek
def c_sigmaclip1D_mean(wave, flux, rms_noise, regions=None, sigma_lower=0.1, sigma_upper=3.0):
    """
    Perform sigma-clipping to determine the mean flux level, with different
    adaptations for emission- and absorption-dominated spectra
    It runs on one-dimensional arrays
    Parameters
    ----------
    wave : np.ndarray
        One-dimension array of wavelength values
    flux : np.ndarray
        One-dimension array of flux values
    rms_noise : float
        The estimated RMS noise level of the data
    line_regions : list
        One-dimension list of [lower_wl, upper_wl] pairs describing regions of significant absorption
    sigma_clip_threshold : float
        The threshold in number of sigma above/below which to reject outlier
        data
    Returns
    -------
    sigmaclip_flux_prev : float
    sigmaclip_flux : float
    sigmaclip_noise : float
        The measured continuum flux and estimated 1-sigma per-channel noise
        around that measurement
        The variable sigmaclip_flux_prev contains the continuum determined
        with the sigma-clipping method without applying the correction
        of STATCONT
    """
    flux_copy = np.copy(flux)
    # Sigma-clipping method applied to the flux array
    if astropy.version.major >= 1:
        if regions is not None:
            for lower_wl, upper_wl in regions:
                mask = (wave>lower_wl) & (wave<upper_wl)
                flux_copy[mask] = np.nan
        if np.isnan(flux_copy).all():
            return None, np.nan, np.nan, np.nan, np.nan
        elif sum(np.isnan(flux_copy)) > 0.5*len(flux_copy):
            return None, np.nan, np.nan, np.nan, np.nan
        filtered_data = sigma_clip(flux_copy, sigma_lower=0.1, sigma_upper=2.0,
                                   iters=1)
        filtered_data = sigma_clip(filtered_data.data, sigma_lower=sigma_lower, sigma_upper=sigma_upper,
                                   iters=None)
    elif astropy.version.major < 1:
        if regions is not None:
            flux_copy = np.copy(flux)                
            
        filtered_data = sigma_clip(flux, sig=sigma_clip_threshold, iters=None)
    
    good_data = filtered_data.data[~filtered_data.mask]
    sigmaclip_flux_prev = sigmaclip_flux = np.median(good_data)
    sigmaclip_noise = sigmaclip_sigma = np.std(filtered_data)
    mean_flux = np.mean(flux)
    
    filt2 = sigma_clip(flux_copy, sigma_lower=5.0, sigma_upper=3.0, iters=1)
    med = np.median(filt2)
    std = np.std(filt2)
    
    sigmaclip_relative_noise = std / med
    # Correction of sigma-clip continuum level, making use of the rms noise
    #sigmaclip_flux = sigmaclip_flux_prev - 1.0*sigmaclip_relative_noise
            
    return sigmaclip_flux_prev, sigmaclip_flux, sigmaclip_noise, filtered_data, sigmaclip_relative_noise
```

```python
# Borrowed from https://github.com/radio-astro-tools/statcont/blob/master/statcont/cont_finding.py
# Modified by Spencer Bialek
def c_sigmaclip1D_top(wave, flux, rms_noise, regions=None, sigma_lower=0.1, sigma_upper=3.0):
    """
    Perform sigma-clipping to determine the mean flux level, with different
    adaptations for emission- and absorption-dominated spectra
    It runs on one-dimensional arrays
    Parameters
    ----------
    wave : np.ndarray
        One-dimension array of wavelength values
    flux : np.ndarray
        One-dimension array of flux values
    rms_noise : float
        The estimated RMS noise level of the data
    line_regions : list
        One-dimension list of [lower_wl, upper_wl] pairs describing regions of significant absorption
    sigma_clip_threshold : float
        The threshold in number of sigma above/below which to reject outlier
        data
    Returns
    -------
    sigmaclip_flux_prev : float
    sigmaclip_flux : float
    sigmaclip_noise : float
        The measured continuum flux and estimated 1-sigma per-channel noise
        around that measurement
        The variable sigmaclip_flux_prev contains the continuum determined
        with the sigma-clipping method without applying the correction
        of STATCONT
    """
    flux_copy = np.copy(flux)
    # Sigma-clipping method applied to the flux array
    if astropy.version.major >= 1:
        if regions is not None:
            for lower_wl, upper_wl in regions:
                mask = (wave>lower_wl) & (wave<upper_wl)
                flux_copy[mask] = np.nan
        if np.isnan(flux_copy).all():
            return None, np.nan, np.nan, np.nan, np.nan
        elif sum(np.isnan(flux_copy)) > 0.5*len(flux_copy):
            return None, np.nan, np.nan, np.nan, np.nan
        filtered_data = sigma_clip(flux_copy, sigma_lower=0.1, sigma_upper=2.0,
                                   iters=1)
        filtered_data = sigma_clip(filtered_data.data, sigma_lower=sigma_lower, sigma_upper=sigma_upper,
                                   iters=None)
    
    elif astropy.version.major < 1:
        if regions is not None:
            flux_copy = np.copy(flux)                
            
        filtered_data = sigma_clip(flux, sig=sigma_clip_threshold, iters=None)
    
    # Find the 10 highest values and fit a line to them, interpolating the rest of the points
    highest_values_indices = filtered_data.data.argsort()[-50:-10][::-1]
    x = wave[highest_values_indices]
    y = filtered_data[highest_values_indices]
    xvals = wave
    z = np.poly1d(np.polyfit(x, y, 1))
    yinterp = z(xvals)
    
    sigmaclip_flux_prev = sigmaclip_flux = np.mean(filtered_data)
    sigmaclip_noise = sigmaclip_sigma = np.std(filtered_data)
    mean_flux = np.mean(flux)
    
    filt2 = sigma_clip(flux_copy, sigma_lower=5.0, sigma_upper=3.0, iters=1)
    med = np.median(filt2)
    std = np.std(filt2)
    
    sigmaclip_relative_noise = std / med
    
    # Correction of sigma-clip continuum level, making use of the
    # presence of emission and/or absorption line features
    
    # Set up the fraction of channels (in %) that are in emission
    fraction_emission = 0
    fraction_emission = sum(fl > (yinterp[i]+1*rms_noise) for i,fl in enumerate(flux))
    fraction_emission = 100*fraction_emission/len(flux)
    
    # Set up the fraction of channels (in %) that are in absorption
    fraction_absorption = 0
    fraction_absorption = sum(fl < (yinterp[i]-1*rms_noise) for i,fl in enumerate(flux))
    fraction_absorption = 100*fraction_absorption/len(flux)

    # Apply correction to continuum level
    # see details in Sect. 2.4 of Sanchez-Monge et al. (2017)
    if (fraction_emission < 33 and fraction_absorption < 33):
        sigmaclip_flux = sigmaclip_flux_prev
    elif (fraction_emission >= 33 and fraction_absorption < 33):
        if (fraction_emission-fraction_absorption > 25):
            sigmaclip_flux = yinterp - 1.0*rms_noise
        if (fraction_emission-fraction_absorption <= 25):
            sigmaclip_flux = yinterp - 0.5*rms_noise
    elif (fraction_emission < 33 and fraction_absorption >= 33):
        if (fraction_absorption-fraction_emission > 25):
            sigmaclip_flux = yinterp + 1.0*rms_noise
        if (fraction_absorption-fraction_emission <= 25):
            sigmaclip_flux = yinterp + 0.5*rms_noise
    elif (fraction_emission >= 33 and fraction_absorption >= 33):
        if (fraction_emission-fraction_absorption > 25):
            sigmaclip_flux = yinterp - 1.0*rms_noise
        if (fraction_absorption-fraction_emission > 25):
            sigmaclip_flux = yinterp + 1.0*rms_noise
        if (abs(fraction_absorption-fraction_emission) <= 25):
            sigmaclip_flux = yinterp
            
    return sigmaclip_flux_prev, sigmaclip_flux, sigmaclip_noise, sigmaclip_flux, sigmaclip_relative_noise
```

```python
def c_sigmaclip1D_STATCONT(wave, flux, rms_noise, regions=None, sigma_clip_threshold=1.8):
    """
    Perform sigma-clipping to determine the mean flux level, with different
    adaptations for emission- and absorption-dominated spectra
    It runs on one-dimensional arrays
    Parameters
    ----------
    flux : np.ndarray
        One-dimension array of flux values
    rms_noise : float
        The estimated RMS noise level of the data
    betaversion : logic
        Activate more functionalities for developers
    sigma_clip_threshold : float
        The threshold in number of sigma above/below which to reject outlier
        data
    Returns
    -------
    sigmaclip_flux_prev : float
    sigmaclip_flux : float
    sigmaclip_noise : float
        The measured continuum flux and estimated 1-sigma per-channel noise
        around that measurement
        The variable sigmaclip_flux_prev contains the continuum determined
        with the sigma-clipping method without applying the correction
        of STATCONT
    """
    
    flux_copy = np.copy(flux)
    # Sigma-clipping method applied to the flux array
    if astropy.version.major >= 1:
        if regions is not None:
            for lower_wl, upper_wl in regions:
                mask = (wave>lower_wl) & (wave<upper_wl)
                flux_copy[mask] = np.nan
        if np.isnan(flux_copy).all():
            return None, np.nan, np.nan
        elif sum(np.isnan(flux_copy)) > 0.5*len(flux_copy):
            return None, np.nan, np.nan
        #filtered_data = sigma_clip(flux, sigma=sigma_clip_threshold,
        #                           iters=None)
        filtered_data = sigma_clip(flux_copy, sigma_lower=0.5, sigma_upper=2.0,
                                   iters=None)
    elif astropy.version.major < 1:
        filtered_data = sigma_clip(flux, sig=sigma_clip_threshold, iters=None)
        
    sigmaclip_flux_prev = sigmaclip_flux = np.mean(filtered_data)
    sigmaclip_noise = sigmaclip_sigma = np.std(filtered_data)
    mean_flux = np.mean(flux)
    
    # Correction of sigma-clip continuum level, making use of the
    # presence of emission and/or absorption line features
    
    # Set up the fraction of channels (in %) that are in emission
    fraction_emission = 0
    fraction_emission = sum(i > (sigmaclip_flux+1*rms_noise) for i in flux)
    fraction_emission = 100*fraction_emission/len(flux)
    
    # Set up the fraction of channels (in %) that are in absorption
    fraction_absorption = 0
    fraction_absorption = sum(i < (sigmaclip_flux-1*rms_noise) for i in flux)
    fraction_absorption = 100*fraction_absorption/len(flux)

    # Apply correction to continuum level
    # see details in Sect. 2.4 of Sanchez-Monge et al. (2017)
    if (fraction_emission < 33 and fraction_absorption < 33):
        sigmaclip_flux = sigmaclip_flux_prev
    elif (fraction_emission >= 33 and fraction_absorption < 33):
        if (fraction_emission-fraction_absorption > 25):
            sigmaclip_flux = sigmaclip_flux_prev - 1.0*rms_noise
        if (fraction_emission-fraction_absorption <= 25):
            sigmaclip_flux = sigmaclip_flux_prev - 0.5*rms_noise
    elif (fraction_emission < 33 and fraction_absorption >= 33):
        if (fraction_absorption-fraction_emission > 25):
            sigmaclip_flux = sigmaclip_flux_prev - 1.0*rms_noise
        if (fraction_absorption-fraction_emission > 50):
            sigmaclip_flux = sigmaclip_flux_prev - 0.5*rms_noise
        if (fraction_absorption-fraction_emission > 75):
            sigmaclip_flux = sigmaclip_flux_prev - 0.25*rms_noise
        if (fraction_absorption-fraction_emission <= 25):
            sigmaclip_flux = sigmaclip_flux_prev + 0.5*rms_noise
    elif (fraction_emission >= 33 and fraction_absorption >= 33):
        if (fraction_emission-fraction_absorption > 25):
            sigmaclip_flux = sigmaclip_flux_prev - 1.0*rms_noise
        if (fraction_absorption-fraction_emission > 25):
            sigmaclip_flux = sigmaclip_flux_prev + 1.0*rms_noise
        if (abs(fraction_absorption-fraction_emission) <= 25):
            sigmaclip_flux = sigmaclip_flux_prev

 
    return sigmaclip_flux_prev, sigmaclip_flux, sigmaclip_noise
```

```python
def fit_continuum(wav, flux, segments, line_regions, sigma_upper=2.0, sigma_lower=0.5, rms_noise=0.1, fit='mean'):
    
    cont_array = np.ones(len(wav))
    rel_noise_array = []
    cont_prev_array = np.ones(len(wav))
    
    for k in range(len(segments)):
        if k == range(len(segments))[-1]:
            if fit=='top_vals':
                cont_prev, cont_, _, _, rel_noise = c_sigmaclip1D_top(wave=wav[segments[k]:], 
                                                      flux=flux[segments[k]:], 
                                                      rms_noise=rms_noise,
                                                      sigma_upper=sigma_upper,
                                                      sigma_lower=sigma_lower,
                                                      regions=line_regions)
            elif fit=='mean':
                cont_prev, cont_, _, _, rel_noise = c_sigmaclip1D_mean(wave=wav[segments[k]:], 
                                                      flux=flux[segments[k]:], 
                                                      rms_noise=rms_noise,
                                                      sigma_upper=sigma_upper,
                                                      sigma_lower=sigma_lower,
                                                      regions=line_regions)
            elif fit=='statcont':
                cont_prev, cont_, rel_noise = c_sigmaclip1D_STATCONT(wave=wav[segments[k]:], 
                                                      flux=flux[segments[k]:], 
                                                      rms_noise=rms_noise,
                                                      regions=line_regions)
            cont_array[segments[k]:] = cont_
            cont_prev_array[segments[k]:] = cont_prev
            rel_noise_array.append(rel_noise)
            
            
        else:
            if fit=='top_vals':
                cont_prev, cont_, _, _, rel_noise  = c_sigmaclip1D_top(wave=wav[segments[k]:segments[k+1]],
                                                       flux=flux[segments[k]:segments[k+1]],
                                                       rms_noise=rms_noise,
                                                       sigma_upper=sigma_upper,
                                                       sigma_lower=sigma_lower,
                                                       regions=line_regions)
            elif fit=='mean':
                cont_prev, cont_, _, _, rel_noise  = c_sigmaclip1D_mean(wave=wav[segments[k]:segments[k+1]],
                                                       flux=flux[segments[k]:segments[k+1]],
                                                       rms_noise=rms_noise,
                                                       sigma_upper=sigma_upper,
                                                       sigma_lower=sigma_lower,
                                                       regions=line_regions)
            elif fit=='statcont':
                cont_prev, cont_, rel_noise  = c_sigmaclip1D_STATCONT(wave=wav[segments[k]:segments[k+1]],
                                                       flux=flux[segments[k]:segments[k+1]],
                                                       rms_noise=rms_noise,
                                                       regions=line_regions)
            cont_array[segments[k]:segments[k+1]] = cont_
            cont_prev_array[segments[k]:segments[k+1]] = cont_prev
            rel_noise_array.append(rel_noise)

    prev_val=0
    beg_idx=0
    end_idx=0
    for i, val in enumerate(cont_array):
        if np.isnan(val) and not np.isnan(prev_val):
            beg_idx = i-1
            beg_val = prev_val
        elif not np.isnan(val) and np.isnan(prev_val):
            end_idx = i
            end_val = val

            # Interpolate between values on either side of the segment of NaNs
            x = [wav[beg_idx], wav[end_idx]]
            y = [beg_val, end_val]
            xvals = wav[beg_idx:end_idx]
            yinterp = np.interp(xvals, x, y)

            # Patch back into cont_array
            cont_array[beg_idx:end_idx] = yinterp
        prev_val = val
        
    prev_val=0
    beg_idx=0
    end_idx=0
    for i, val in enumerate(cont_prev_array):
        if np.isnan(val) and not np.isnan(prev_val):
            beg_idx = i-1
            beg_val = prev_val
        elif not np.isnan(val) and np.isnan(prev_val):
            end_idx = i
            end_val = val

            # Interpolate between values on either side of the segment of NaNs
            x = [wav[beg_idx], wav[end_idx]]
            y = [beg_val, end_val]
            xvals = wav[beg_idx:end_idx]
            yinterp = np.interp(xvals, x, y)

            # Patch back into cont_array
            cont_prev_array[beg_idx:end_idx] = yinterp
        prev_val = val
        
    # Adjust continuum based on relative noise of spectrum
    rel_noise = np.nanmean(rel_noise_array)
    #cont_array_adjusted = cont_array - 100*rel_noise
    
    return cont_array, cont_prev_array, rel_noise#cont_array_adjusted
```

```python
def get_rms_noise(wave, flux, regions):
    
    noise_array = []
    for lower_wl, upper_wl in regions:
        mask_ = (wave>lower_wl) & (wave<upper_wl)
        flux_ = flux[mask_]
        wave_ = wave[mask_]
        
        filtered_data = sigma_clip(flux_, sigma_lower=2., sigma_upper=2.,
                                   iters=None)
        good_data = filtered_data.data[~filtered_data.mask]
        
        signal = np.median(good_data)
        noise = np.std(good_data)
        
        noise_array.append(noise)
    
    return np.mean(noise_array)
```
